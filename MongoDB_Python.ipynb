```python
#mise a jour d'un champ dans un element d'une collection 
publications.update_one({ "title": "Post Title 1" }, { "$set": { "likes": 2 } } )
pprint.pprint(publications.find_one({ "title": "Post Title 1"})) 
```

```python
#Mise a jouter du champ likes en l'incrementant à 1 dans tous les élément 
#publications.update_many({}, { "$inc": { "likes": 1 } })
for publication in publications.find():
  pprint.pprint(publication)
```

```python
#Ajout d'un nouvel tags au titre nommés : Post Title 1 
publications.update_one({ "title": "Post Title 1" }, { "$push": {"tags" : "Tech" }  } )
pprint.pprint(publications.find_one({ "title": "Post Title 1"})) 
```

```python
#Suppression d'un nouvel tags au titre nommés : Post Title 1 
publications.update_one({ "title": "Post Title 1" }, { "$pull": {"tags" : "Tech" }  } )
pprint.pprint(publications.find_one({ "title": "Post Title 1"})) 
```

```python
#Suppression d'un element de la collection publication 
publications.delete_one({"author": "Mounia"})
for publication in publications.find():
  pprint.pprint(publication)
```

```python
#Affichage des categorie de publication et leur nombre de like 
rest=publications.aggregate([
  {
    "$match": { "likes": { "$gt": 1 } }
  },
  {
    "$group": { "_id": "$category", "totalLikes": { "$sum": "$likes" } }
  }
])
for r in rest:
    pprint.pprint(r)
```

```python
#Affichage des publication qui sont sortie avant le 12 novembre 2009 à 12:00 et les ordonner par nom d'auteur 
d = datetime.datetime(2009, 11, 12, 12)
for publication in publications.find({"date": {"$lt": d}}).sort("author"):
  pprint.pprint(publication)
```

```python
#Rajouter des elements dans la collection publication 
import random 
for i in range(10000):
    author="AUTHOR_"
    author += str(i)
    body="Body of post."
    category = "Technology"
    date = datetime.datetime.utcnow()
    likes = int(random.randint(1, 100))
    tags = ["news"]
    title="Post Title "
    title += str(i)
    publications.insert_one({"author":author, "body":body, "category":category, "date":date,"likes":likes,"tags":tags,"title":title})
```
